## 实验一 

```python
import pandas as pd
#来自维基百科的所有奥运会奖牌，用01！02！03！来表示金、银、铜
df = pd.read_csv('olympics.csv')
df.head()
```

```python
#设置行的标签从第二行数据开始读取，列的索引从第二列开始
df = pd.read_csv('olympics.csv', index_col = 0, skiprows=1)
df.head()
```

```python
df.columns
```

```python
for col in df.columns:
    if col[:2]=='01':
        df.rename(columns={col:'Gold' + col[4:]}, inplace=True)
    if col[:2]=='02':
        df.rename(columns={col:'Silver' + col[4:]}, inplace=True)
    if col[:2]=='03':
        df.rename(columns={col:'Bronze' + col[4:]}, inplace=True)
    if col[:1]=='?':
        df.rename(columns={col:'#' + col[1:]}, inplace=True) 

df.head()
```

### Q1-编写代码将原df更改为以下的格式（修改行索引的名称，新增一列‘ID’存储个国家的缩写）并删除行索引名为Totals的行 

```python
names_ids = df.index.str.split('\s\(') # split the index by '('

df.index = names_ids.str[0] # the [0] element is the country name -> new index 
df['ID'] = names_ids.str[1].str[:3] # the [1] element is the abbreviation or ID (take first 3 characters from that)
df=df.drop('Totals') 
df.head()
```

### Q2-获得国家名为Armenia的所有夏季奥运会的获奖信息 

```python
df.loc['Armenia'][:5]
```

### Q3-冬季获得金牌数量大于银牌数量的国家的所有信息 

```python
df[df['Gold.1']>df['Silver.1']]
```

### Q4-获取[Gold:Total]两列以及中间的所有数据，采用连续切片操作 

```python
df.iloc[:,1:5]
```

### Q5-哪个国家在夏季奥运会上获得了最多的金牌？

```python
pdf=df.sort_values(by='Gold',ascending=False)
pdf.iloc[0].name
```

## 实验二

```python
from pandas_datareader import data
import matplotlib.pyplot as plt
import seaborn
%matplotlib inline
seaborn.set()
google = data.DataReader('GOOG', start='2005', end='2012',
                       data_source='yahoo')
google.head()
```

```python
google = google['Open']
google.head(10)
```

### Q1-对开盘价以周为频率采样，并以左标签、右闭合的样式返回均值

```python
google.resample('W', closed='right').mean()
```

### Q2-以天为频率采样，使用每周五的开盘价填充周末的缺失值 

```python
google.resample('D').ffill()
```

### Q3-采样每个月末的开盘价 

```python
google.asfreq('M')
```

### Q4-使开盘价格向前偏移3个工作日（即当天对应三个工作日后的开盘价）

```python
google.shift(-3)
```
