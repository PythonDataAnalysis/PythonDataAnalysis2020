## Q2-利用随机函数，随机生成范围在0到50之间到100个数据，并绘制频次直方图
## 要求：均匀10个区间，透明度为0.8，xlabel为"data",ylabel为"count"

```python
data = np.random.randint(0,50,100)
bins=np.arange(0,51,5) 

plt.hist(data,bins,alpha=0.8)
plt.xlabel('data')
plt.ylabel('count')
plt.show()
```

## Q3-补全代码，生成如下饼图

```python
labels = 'A', 'B', 'C', 'D', 'others'
sizes = [10.4, 20.6, 21.45, 35, 12.55]
explode = (0, 0, 0, 0, 0)
plt.pie(sizes, explode=explode, labels=labels,
        autopct='%1.2f%%', shadow=False, startangle=90)
plt.show()
```

## Q4-补全代码，生成如下极坐标，要求：采样点数为45

```python
a = plt.subplot(121, projection = 'polar')
b = plt.subplot(122, projection = 'polar')

t1 = np.linspace(0, 2*np.pi, 30)
t2 = np.linspace(0, 2*np.pi, 45)
a.plot(t1, np.sin(3*t1), '.', c='g')
b.plot(t2, np.sin(3*t2), 'x', c='b')
plt.show()
```

## Q5-补全代码，生成如下三维图，并将俯仰角调整为 60 度，方位角调整为 45 度

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = Axes3D(fig)
X = np.arange(-3, 3, 0.1)
Y = np.arange(-3, 2, 0.1)
X,Y = np.meshgrid(X,Y)
def f(x,y):
    return (1-y**5+x**5)*np.exp(-x**2-y**2)
ax.plot_surface(X,Y,f(X,Y),rstride=1,cstride=1)

plt.show()
```
